```python
import numpy as np
import matplotlib.pyplot as plt
```

# Generate Data

```python
jumlah_data = 100

y = np.array([i*0.1+np.random.randn() for i in range(jumlah_data)])
x = np.array([i*0.1 for i in range(jumlah_data)])
```

### Plot Data

### Funsi Regresi Linear

```python
def regrsi_linear(x,gradient):
    y = gradient*x
    return y
```

menghitung persamaan linear

```python
x_prediction = np.array([0,10])
m_prediction = 5
y_prediction = regrsi_linear(x_prediction,m_prediction)
```

```python
plt.scatter(x, y)
plt.plot(x_prediction,y_prediction, "Green")
plt.axis([0,10,0,10])
plt.show()
```
